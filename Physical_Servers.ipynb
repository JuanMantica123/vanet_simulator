```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import vanet_tools as vp
```

```python
micro_df = pd.read_csv('manhattan_config.csv',index_col=0).values
```

```python
p_v_ratio, num_virtual = 1/5,len(micro_df)
num_physical = int(p_v_ratio*num_virtual)
```

```python
kmeans = KMeans(n_clusters=num_physical).fit(micro_df)
y_kmeans = kmeans.predict(micro_df)
```

```python
plt.scatter(micro_df[:, 0], micro_df[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
ks = [k for k in range(1,len(micro_df)+1)]
sum_of_squared_distances = []
for k in ks:
    kmeans = KMeans(n_clusters=k).fit(micro_df)
    sum_of_squared_distances.append(kmeans.inertia_)
```

```python
plt.plot(ks,sum_of_squared_distances)
plt.show()
```

```python
pasubio_df = pd.read_csv('Pasubio_config.csv',index_col=0).values
```

```python
ks = [k for k in range(1,len(pasubio_df)+1)]
sum_of_squared_distances = []
for k in ks:
    kmeans = KMeans(n_clusters=k).fit(micro_df)
    sum_of_squared_distances.append(kmeans.inertia_)
```

```python
plt.plot(ks,sum_of_squared_distances)
plt.show()
```

```python
num_physical_pasubio = 5
kmeans = KMeans(n_clusters=num_physical_pasubio).fit(pasubio_df)
y_kmeans = kmeans.predict(pasubio_df)
centers = kmeans.cluster_centers_
```

```python
plt.scatter(pasubio_df[:, 0], pasubio_df[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
physical_server_range = 300
```

```python
fig, ax = plt.subplots()
plt.xlabel("x coordinate")
plt.ylabel("y coordinate")
plt.title("Microclouds Under RSU Range")
for physical_server_coords in centers:
    physical_server = plt.Circle(physical_server_coords,physical_server_range)
    ax.add_artist(physical_server)
plt.scatter(pasubio_df[:, 0], pasubio_df[:, 1], c=y_kmeans, s=100, cmap='viridis',zorder=10)
plt.show()
```
